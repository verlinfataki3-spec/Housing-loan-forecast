## 6. Détection & gestion des outliers

- Méthode IQR ou Z-score  
- Décision métier : supprimer / winsoriser / transformer (log)

```python
# Detection des outliers avec la methode IQR

def detect_outliers_iqr(df, k=1.5):
    summary = []
    details = {}
    
    num_cols = df.select_dtypes(include=[np.number]).columns
    
    for col in num_cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower, upper = Q1 - k*IQR, Q3 +k*IQR
        
        mask = (df[col] < lower) | (df[col] > upper)
        outliers_idx = df[mask].index
        
        details[col] = outliers_idx
        summary.append({
            "colonne": col,
            "nb_outliers": mask.sum(),
            "pct_outliers": round(100*mask.mean(), 2)
        })
    summary = pd.DataFrame(summary).sort_values("pct_outliers", ascending=False)
    return summary, details

# === Exemple d'utilisation ===
summary, details = detect_outliers_iqr(df)

print("Résumé par colonne :")
display(summary)

print("\nIndices des outliers pour la colonne 'Glucose' (exemple) :")
print(details.get("Glucose", []))   

    
```

```python
# Detection des outliers avec la methode Z-score
def detect_outliers_zscore(df, threshold=3.0):
    rows = []
    details = {}
    for col in df.select_dtypes(include=[np.number]).columns:
        mean, std = df[col].mean(), df[col].std()
        if std == 0 or np.isnan(std):
            continue
        z = (df[col] - mean) / std
        mask = z.abs() > threshold
        details[col] = df.index[mask]
        rows.append({
            "colonne": col, 
            "nb_outliers_z": int(mask.sum()), 
            "pct_outliers_z": round(100*mask.mean(), 2)
        })
    rows = pd.DataFrame(rows).sort_values("pct_outliers_z", ascending=False)
    return rows, details

rows, details = detect_outliers_zscore(df)


print("Résumé par colonne :")
display(rows)

print("\nIndices des outliers pour la colonne 'Glucose' (exemple) :")
print(details.get("Glucose", [])) 
    
```
