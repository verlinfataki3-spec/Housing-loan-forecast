```python
# Separations des variables categoriques et numeriques
df_num = []
df_cat = []

for col in df.columns:
    if pd.api.types.is_numeric_dtype(df[col]):
        df_num.append(col)
    else:
        df_cat.append(col)
```

```python
# Verification de nos variables numeriques
df_num
```

```python
# Verification de nos variables categoriques
df_cat
```

## 5. Nettoyage & Feature Engineering
- Imputation simple des NA (médiane/mode):
    - Pour les variables categoriques: on va les remplacer par le mode
    - Pour les variables numeriques: on va les remplacer par la mediane

```python
# Remplacer les valeurs manquantes pour les variables categoriques par la mode
for col in df_cat:
    mode_series = df[col].mode(dropna=True)
    if not mode_series.empty:
        mode_val = mode_series[0]
        df[col] = df[col].fillna(mode_val)
```

```python
# Remplacer les valeurs manquantes pour les variables numeriques par la mediane
for col in df_num:
    median_series = df[col].median()
    df[col] = df[col].fillna(median_series)
```

```python
# La verification de du dataset pour voir s'il n'y a plus des valeurs manquantes
df.isnull().sum()
```
